```python
symbol_counts = df.groupby('Symbol', as_index=False).size().rename(columns={'size': 'row_count'}).sort_values(
    'row_count', ascending=False)
min_row_count = symbol_counts['row_count'].min()
min_row_count
```

```python
latest_dates = pd.to_datetime(df['Date']).sort_values().unique()[-min_row_count:]
filtered = df[df['Date'].isin(latest_dates)]
latest_weighted_table = filtered.pivot_table(index='Date', columns='Symbol', values='Weighted_Return_10d',
                                             aggfunc='last').sort_index()
latest_weighted_table
```
