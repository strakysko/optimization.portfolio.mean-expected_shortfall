Let's check the consistency between tickers in both datasets by:
1. Loading the sp500_daily_close data
2. Comparing tickers from both datasets

```python
sp500_daily_close = pd.read_parquet('../data/raw/sp500_daily_close.parquet')
tickers_close = set(sp500_daily_close.Symbol)
tickers_weights = set(sp500_weights['Ticker'])

print(f"Number of tickers in daily close: {len(tickers_close)}")
print(f"Number of tickers in weights: {len(tickers_weights)}")
print(f"\nTickers in weights but not in close data: {tickers_weights - tickers_close}")
print(f"\nTickers in close data but not in weights: {tickers_close - tickers_weights}")
```
